```python
df = df.drop(['id'], axis=1)
```

```python
df.head()
```

```python
cat_cols = [
    'gender',
    'ethnicity',
    'smoking_status',
    'employment_status'
]
df = pd.get_dummies(
    df,
    columns=cat_cols,
    drop_first=True   # avoids dummy variable trap
)
```

```python
df.head()
```

```python
df.info()
```

```python
df['bp_category'] = (
    (df['systolic_bp'] >= 140) | (df['diastolic_bp'] >= 90)
).astype(int)
```

```python
df.head()
```

```python
df['obese'] = (df['bmi'] >= 30).astype(int)
```

```python
df['sleep_risk'] = (
    (df['sleep_hours_per_day'] < 6) | 
    (df['sleep_hours_per_day'] > 9)
).astype(int)
```

```python
df['cholesterol_ratio'] = df['ldl_cholesterol'] / df['hdl_cholesterol']
```

```python
df.head()
```

```python
df['activity_level'] = pd.cut(
    df['physical_activity_minutes_per_week'],
    bins=[-1, 149, 300, np.inf],
    labels=[0, 1, 2]
).astype(int)
```

```python
df['lifestyle_risk'] = (
    (df.filter(like='smoking_status_').sum(axis=1) > 0).astype(int) +
    (df['alcohol_consumption_per_week'] > 14).astype(int) +
    (df['screen_time_hours_per_day'] > 6).astype(int) +
    (df['activity_level'] == 0).astype(int)
)

```

```python
df.info()
```

```python
df['metabolic_risk'] = (
    (df['bmi'] >= 30).astype(int) +
    (df['bp_category'] == 1).astype(int) +
    (df['cholesterol_ratio'] > 5).astype(int) +
    (df['triglycerides'] > 150).astype(int)
)
```

```python
df[
    ['bp_category', 'cholesterol_ratio', 'obese',
     'activity_level', 'sleep_risk',
     'lifestyle_risk', 'metabolic_risk']
].head()
```

```python
df.info()
```

```python
education_map = {
    'No schooling': 0,
    'High school': 1,
    'Graduate': 2,
    'Postgraduate': 3
}

income_map = {
    'Low': 0,
    'Medium': 1,
    'High': 2
}

df['education_level'] = df['education_level'].map(education_map)
df['income_level'] = df['income_level'].map(income_map)
```

```python
df.select_dtypes(include='object').columns
```

```python
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
df.info()
```

```python
df['education_level'] = df['education_level'].fillna(df['education_level'].median())
df['income_level'] = df['income_level'].fillna(df['income_level'].median())

```
